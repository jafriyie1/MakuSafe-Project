```python
#Read in file

file = open("upstairsAll.csv")
fileInput =  file.read()
```

```python
#remove endline characters
fileFixed = ''
for line in fileInput:
    fileFixed += line.replace('\n', ',')
```

```python
#split file into a list
fileSplit = [x.strip() for x in fileFixed.split(',')]
#print(fileSplit[1])
```

```python
#format list of list
fileSplitFixed = []
myList = []
for x in range(len(fileSplit)):
    if x%4 == 0:
        fileSplitFixed.append(myList)
        myList = []
    else:
        myList.append(fileSplit[x])
```
